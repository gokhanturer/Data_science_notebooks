### plt.tight_layout() -->  grafikteki anomalileri regule ediyor (otomatik adjust ediyor)

```python
# functional

age = [25, 27, 29, 31, 33, 35, 37, 39, 41, 43, 45]
salary = [38496, 42000, 46752, 49320, 53200, 56000, 62316, 64928, 67317, 68748, 73752]
salary_2 = [45372, 48876, 53850, 57287, 63016,
            65998, 70003, 70000, 71496, 75370, 83640]

plt.subplot(2,1,1) # 2 satır 1 sütun 1. indeks. 2 satırlı 1 sütunlu bir plotun 1.si 
plt.plot(age, salary, "r")
plt.xlabel("age vs salary1")

plt.subplot(2,1,2) # 2 satır 1 sütun 2. indeks. 2 satırlı 1 sütunlu bir plotun 2.si 
plt.plot(age, salary_2, "b")
plt.xlabel("age vs salary2")

plt.tight_layout()
plt.show()
```

```python
# functional

plt.subplot(1,2,1)   # 1 satır 2 sütun olacak. ve altında belirtilen age, salary axesi bunun 1.indeksi olacak.
plt.plot(age, salary, "r")
plt.xlabel("age vs salary1")

plt.subplot(1,2,2)  # 1 satır 2 sütun olacak. ve altında belirtilen age, salary_2 axesi bunun 2.indeksi olacak.
plt.plot(age, salary_2, "b")
plt.xlabel("age vs salary2")
plt.tight_layout()
plt.show()
```

## object oriented (OOP) metodu ile yaparsak

```python
# yukardaki işlemi object oriented metodu ile kolayca aşağıdaki gibi yapabiliriz.

fig, ax = plt.subplots(2,3)
plt.tight_layout()

ax[1,1].plot(x,y)  # 1. satır, 1. sütundaki subplotu bu şekilde yap
ax[0,2].plot(x,x**2) # 0. satır, 2. sütundaki subplotu bu şekilde yap
plt.show()
```

```python
plt.subplot(2,2,1)
plt.plot(age, salary, "r+")  # red renkte ve + ile markala
plt.subplot(2,2,2)
plt.plot(age, salary, "b--") # blue renkte ve -- ile markala
plt.subplot(2,2,3)
plt.plot(age, salary, "mo") # 
plt.subplot(2,2,4)
plt.plot(age, salary, "c.") 
```

## color attribute

```python
fig, ax = plt.subplots(figsize=(6,4))
ax.plot(x,y, color ="red")
ax.plot(x,y-1, color ="blue")
ax.plot(x,y-2, color ="yellow")
ax.plot(x,y-3, color ="black")
```

## linewidth (lw) attribute

```python
fig, ax = plt.subplots(figsize=(6,4))
ax.plot(x,y, color ="red", linewidth = 5)  # kırmızı çizginin genişliğini ayarlıyorum
ax.plot(x,y-1, color ="blue", lw = 4)  # lw olarak kısaltarak mavinin genişliğini ayarladım.
ax.plot(x,y-2, color ="yellow", lw = 3) # sarının genişliğini 3 yaptım
ax.plot(x,y-3, color ="black", lw = 0.5) # siyahın genişliği 0.5 oldu
```

## linestyle (la) attribute

```python
fig, ax = plt.subplots(figsize=(6,4))
ax.plot(x,y, color ="red", lw=2, ls = "--") # ls ile linestyle'ı -- olarak belirledim.
ax.plot(x,y-1, color ="blue", lw=2, ls = "-")
ax.plot(x,y-2, color ="yellow", lw=2, ls = "-.")
ax.plot(x,y-3, color ="black", lw=2, ls = ":")
```

## marker attribute

```python
# marker, x ile ye değerlerinin kesiştiği noktalardır. bu noktaları markalıyoruz

fig, ax = plt.subplots(figsize=(6,4))
ax.plot(x,y, marker="+")
ax.plot(x,y-1, )
ax.plot(x,y-2, )
ax.plot(x,y-3, )
```

```python
fig, ax = plt.subplots(figsize=(6,4))
ax.plot(x,y, marker="+", markersize=20)
ax.plot(x,y-1, )
ax.plot(x,y-2, )
ax.plot(x,y-3, )
```

```python
fig, ax = plt.subplots(figsize=(6,4))
ax.plot(x,y, marker="+", markersize=10)
ax.plot(x,y-1, marker="o", markersize=10)
ax.plot(x,y-2, marker="s", markersize=10)
ax.plot(x,y-3, marker="1", markersize=10)
```

```python
x = np.arange(1,10)
fig, ax = plt.subplots(nrows=1,ncols=2, figsize=(10,4))
ax[0].plot(x,x**2,x,x**3,x,x**4)
ax[1].plot(x,x**2,x,x**3,x,x**4)
ax[1].set_xlim([2,4])  # x ekseninde 2 ile 4 aralığını göster (zoom yapıyoruz)
ax[1].set_ylim([9,100]) # y ekseninde 9 ile 100 aralığını göster
```

```python
x = np.arange(1,10)
fig, ax = plt.subplots(nrows=1,ncols=2, figsize=(10,4))

ax[0].plot(x,x**2, x,x**3, x,x**4)
ax[1].plot(x,x**2, x,x**3, x,x**4)

ax[1].set_xlim([2,4])
ax[1].set_ylim([5,100])  

# bunu seçerken xlimitini göz önünde bulundur.
# şekilde x'in 2 ile 4 arasında olan bölgesine karşılık gelen y değeri aşağı yukarı kaç olabilir diye düşüneceğiz.
```

```python
x = np.arange(1,10)
fig, ax = plt.subplots(nrows=1,ncols=2, figsize=(10,4))
ax[0].plot(x,x**2, x,x**3, x,x**4)
ax[1].plot(x,x**2, x,x**3, x,x**4)

ax[1].set_xlim([2,4])
ax[1].set_ylim([5,100])

ax[0].axvline(x=3,ls="--")

# x=3 olduğu yere kadar olan  bölgeyi xlimiti 2-4, ylimiti 5-100 olacak şekilde zoom yaptık.
```

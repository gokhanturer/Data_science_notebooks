```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import scipy
import warnings
warnings.filterwarnings('ignore')
import seaborn as sns
import numpy as np
print(sns.get_dataset_names())
```

```python
penguins = sns.load_dataset("penguins")
penguins
```

```python
penguins.describe()
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(penguins.isnull())
```

```python
sns.catplot(data = penguins, x="species", y="bill_length_mm", kind="bar", estimator=np.count_nonzero, ci="sd")
```

```python

sns.catplot(data = penguins, x="species", y="bill_length_mm", kind="bar", estimator=np.count_nonzero, ci="sd", col="island", hue="sex")
```

```python

sns.catplot(data = penguins, x="species", y="bill_length_mm", kind="box",  ci="sd", col="species", hue="sex")
```

```python
sns.jointplot(data=penguins, x="bill_depth_mm", y="bill_length_mm")
```

```python
sns.jointplot(data=penguins, x="bill_depth_mm", y="bill_length_mm", hue="species")
```

```python
sns.jointplot(data=penguins, x="body_mass_g", y="flipper_length_mm", hue="species")
```

```python
sns.jointplot(data=penguins, x="body_mass_g", y="flipper_length_mm", kind= "hex")
```

```python
sns.jointplot(data=penguins, x="body_mass_g", y="flipper_length_mm",hue="species", kind= "kde")
```

# Pairgrid

```python
g=sns.PairGrid(penguins,hue="species")  # hangi plotsa onu yazdÄ±k
g.map_diag(sns.histplot)
```

```python
g=sns.PairGrid(penguins,hue="species")  # hangi plotsa onu yazdÄ±k
g.map_diag(sns.histplot)
```

```python
g = sns.PairGrid(penguins)
g.map_diag(sns.histplot)
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot)
```

```python
sns.set_palette(["green","blue","red"])
g = sns.PairGrid(penguins, hue = "species")
g.map_upper(sns.scatterplot);
g.map_diag(sns.histplot);
g.map_lower(sns.kdeplot);
g.add_legend();
```

# FacetGrid

```python
g = sns.FacetGrid(data=penguins, col="species", row="island", hue="sex")
g.map(sns.histplot, "bill_length_mm")
```

```python
g = sns.FacetGrid(data = penguins, col = "species", row = "island", hue = "sex")
g.map(sns.kdeplot, "bill_length_mm");
```

```python
g = sns.FacetGrid(data = penguins, col = "species", row = "island", hue = "sex")
g.map(sns.scatterplot, "bill_length_mm", "body_mass_g");
```
